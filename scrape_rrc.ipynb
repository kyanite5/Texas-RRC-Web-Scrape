```python
from splinter import Browser
import pandas as pd
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#url for Injection/Disposal Permit Query Results
url = 'http://webapps2.rrc.texas.gov/EWA/uicQueryAction.do'
browser.visit(url)
```

```python
#finds district 10
district = browser.find_by_name('searchArgs.districtCodeArg').first
district.select('10')
```

```python
#finds salt water as permitted fluid
fluid = browser.find_by_name('searchArgs.permittedFluidArg').first
fluid.select('Salt Water')
```

```python
#selects disposal into active zone
active = browser.find_by_name('searchArgs.injectionTypeArg').first
active.select('Disposal into a productive zone (H-1)')
```

```python
#submits query form
submit = browser.find_by_xpath('/html/body/table[2]/tbody/tr/td/form/table/tbody/tr[2]/td/table/tbody/tr[4]/td/input[1]').click()
```

```python
#downloads results as csv file, saves locally
download = browser.find_by_xpath('/html/body/table[2]/tbody/tr/td/form/table/tbody/tr[2]/td/table/tbody/tr[3]/td/table/tbody/tr/td[1]/input').click()
```

```python
returns = browser.find_by_xpath('/html/body/table[2]/tbody/tr/td/form/table/tbody/tr[2]/td/table/tbody/tr[3]/td/table/tbody/tr/td[2]/input').click()
```
